```python
import pandas as pd
from bokeh.sampledata.us_counties import data as counties
from bokeh.resources import CDN
from bokeh.embed import file_html
```

```python
jobs = pd.read_csv('data/raw_sources/fourth_quarter_employment_by_county.csv', thousands=',')
housing = pd.read_csv('data/raw_sources/ACSDP5Y2022.DP04-Data.csv')
housing = housing.rename(columns={'Estimate!!HOUSING OCCUPANCY!!Total housing units':'housing_units'})
housing
jobs['Geographic Area Name'] = jobs['County']
jobs = jobs.rename(columns={'December Employment':'dec_2023_employment'})
df = jobs.merge(housing,on='Geographic Area Name')
#wage = []
#for index, row in df.iterrows():
#    w = float(row['Average weekly wage(1)'].replace(',',''))
#    wage.append(w)
#df['avg_weekly_wage'] = wage
#df = df.drop(columns='Average weekly wage(1)')
df 
```

```python
counties = {
    code: county for code, county in counties.items()
}

county_xs = [county["lons"] for county in counties.values()]
county_ys = [county["lats"] for county in counties.values()]

county_names = [county['detailed name'] for county in counties.values()]

lonlat = pd.DataFrame(list(zip(county_names, county_xs, county_ys)), columns =['Geographic Area Name', 'lat', 'lon']) 

df = df.merge(lonlat,on='Geographic Area Name')
df['unit_per_job'] = df['housing_units']/df['dec_2023_employment']
df
```

```python
df['unit_per_job'] = round(df['unit_per_job'],2) 
df.to_csv('data\jobs_housing_clean.csv') #save the full thing
df = df[['Geographic Area Name',
         'unit_per_job',
         'dec_2023_employment',
         'housing_units',
         'lat',
         'lon']]
df
```

```python
#trying indiana only

df['State'] = df['Geographic Area Name'].str.split(',').str[1]
dff = df[df['State']==' Indiana']
dff
```

```python
from bokeh.io import show
from bokeh.models import LogColorMapper, GeoJSONDataSource, ColorBar, NumeralTickFormatter, HoverTool
from bokeh.palettes import YlOrRd9 as palette
from bokeh.plotting import figure, show
from bokeh.models import GeoJSONDataSource, ColumnDataSource
from shapely.geometry import Polygon
import json

#create geojson
def create_geojson(df):
    geojson_data = []
    for index, row in df.iterrows():
        coords = list(zip(row['lon'], row['lat']))
        polygon = Polygon(coords)
        geojson_data.append({'Geographic Area Name': row['Geographic Area Name'],
                             'unit_per_job': row['unit_per_job'],
                             'dec_2023_employment': row['dec_2023_employment'],
                             'housing_units': row['housing_units'],
                             'geometry': {'type': 'Polygon', 'coordinates': [list(polygon.exterior.coords)]}})
    return geojson_data

# Convert DataFrame to GeoJSON format
geojson_data = create_geojson(dff)
# Convert the list of dictionaries to a GeoJSON string
geojson_str = json.dumps(geojson_data)
# Create the GeoJSONDataSource
geojson_source = GeoJSONDataSource(geojson=geojson_str)
```

```python
# Set the format of the colorbar
min_range = min(dff['unit_per_job'])
max_range = max(dff['unit_per_job'])

# Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LogColorMapper(palette = palette, low = min_range, high = max_range)

# Create color bar.
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=18,
border_line_color=None, location = (0, 0))
hover = HoverTool(tooltips=[("Name", "@Geographic Area Name"),
                            ("Housing Units Per Job", "@unit_per_job"),
                            ("Employment (Dec 2023)", "@dec_2023_employment"),
                            ("Housing Units (2024)", "@housing_units")
                            ]
                 )
# Create figure object.
p = figure(title = "Occupied Housing Units per Job", 
         plot_height = 650, plot_width = 850,
         toolbar_location = None)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.axis.visible = False

# Add patch renderer to figure. 
p.patches('lat','lon', source = dff, 
          fill_color = {'field' : 'unit_per_job', 'transform' : color_mapper},
          line_color = 'white', line_width = 0.5, fill_alpha = 0.7)

# Specify color bar layout.
p.add_layout(color_bar, 'right')

# Add the hover tool to the graph
p.add_tools(hover)
```

```python
#make map
#palette.reverse()
color_mapper = LogColorMapper(palette=palette)

TOOLS = "pan,wheel_zoom,reset,hover,save"

p = figure(plot_width=800, plot_height=800,
    title="Occupied Housing Units per Job", tools=TOOLS,
    x_axis_location=None, y_axis_location=None,
    tooltips=[
        ("Name", "@Geographic Area Name")
       ,("Housing Units Per Job", "@unit_per_job")
       ,("Employment (Dec 2023)", "@dec_2023_employment")
       ,("Housing Units (2024)", "@housing_units")
             ]
          )

p.grid.grid_line_color = None
p.hover.point_policy = "follow_mouse"

p.patches('xs', 'ys', source=geojson_source,
          fill_color={'field': 'unit_per_job', 'transform': color_mapper},
          fill_alpha=0.7, line_color="white", line_width=0.5)
```

```python
show(p)
```

```python
#trying folium
import folium

# Create a base map centered on the United States
m = folium.Map(location=[41.67, -86.25], zoom_start=5)

for index, row in df.iterrows():
    # Create a GeoJSON feature for each row
    geojson_data = {
        "type": "Feature",
        "geometry": {"type": "Polygon", "coordinates": [list(zip(row['lon'], row['lat']))]},
        "properties": {
            "popup": f"Geographic Area Name: {row['Geographic Area Name']}<br>unit_per_job: {row['unit_per_job']}"
        }
    }
    
    # Add the GeoJSON feature to the map
    folium.GeoJson(geojson_data, popup=folium.Popup(max_width=200)).add_to(m)

m
```
